```python
%%writefile p3.cpp
#include <iostream>
#include <string>
#include <functional>
#include <ctime>
template <class T>
class HashEntidad {
private:	
	int key;
	T value;
public:
	HashEntidad(int key, T value):key(key), value(value) {	}
	int getKey() { return key; }
	T getValue() { return value; }
};
using namespace std;
template <class T>
class HashTabla {
private:
	HashEntidad<T> ** tabla;
	int numElementos;
	int TABLE_SIZE;
  function<int(T)> f_hash;
  function<bool(T,T)> f_igual;
public:
	HashTabla(int TABLE_SIZE, function<int(T)> f_hash_, function<bool(T,T)> f_igual_) {
		this->TABLE_SIZE = TABLE_SIZE;
		tabla = new HashEntidad<T> *[TABLE_SIZE];
		for (int i = 0; i < TABLE_SIZE; ++i) {
			tabla[i] = nullptr;
		}
		numElementos = 0;
    f_hash=f_hash_;
    f_igual=f_igual_;
	}
	~HashTabla()
	{
		for (int i = 0; i < TABLE_SIZE; ++i) {
			if (tabla[i] != nullptr) {
				delete tabla[i];
			}
		}
		delete[] tabla;
	}

  int funcHash(T key){
    return f_hash(key);
  }

	//Direccionamiento según Prueba Lineal
	void insertar(T value) {
		//Hash prima
		int base, step, hash;
		//validar si la tabla está llena
		if (numElementos == TABLE_SIZE)return;
		//Función Hash1
		base = funcHash(value)%TABLE_SIZE;
		hash = base;
		//constante para Hash2
		step = 0;
		while (tabla[hash]!=nullptr)
		{
			//Función Hash2
			hash = (base + step) % TABLE_SIZE;
			step++;
		}
		//almacenarlo en la tabla
		tabla[hash] = new HashEntidad<T>(0,value);
		numElementos++;
	}
	int size() {
		return TABLE_SIZE;
	}
	int sizeactual() {
		return numElementos;
	}
	T buscar(T key) {
		int step = 0;
		int i, base;
		i = base = funcHash(key)%TABLE_SIZE; //hash1 es = a hash2 cuando step=0;
		while (true)
		{
			if (tabla[i] == nullptr)break;
			else if(f_igual((tabla[i]->getValue()),key)) {
				return tabla[i]->getValue();
			}
			else step++;
			i = (base + step) % TABLE_SIZE;
		}
	}
};
string to_string_(int num){
    string xd="";
    int digito;
    int n=num;
    while(n != 0) {
      digito = n % 10;
      n /= 10;
      string a="";
      a.push_back(char(digito+48));
      xd.insert(0,a);
    }
    return xd;
}
class Persona{
  private:
    string nombre, apellido;
    int edad;
  public:
    Persona(string nombre, string apellido,int edad):nombre(nombre),apellido(apellido),
    edad(edad) {}
    string getTodo(){
      return nombre+" "+apellido+" "+to_string_(edad);
    }
    int getEdad(){
      return edad;
    }
    string getNombre(){
      return nombre;
    }
    string getApellido(){
      return apellido;
    }
    friend std::ostream& operator << (std::ostream& a, const Persona& data)
  	{
		if (data.nombre != "")
	  		return a << data.nombre<<endl;
	  	else
  			return a << "NOT FOUND" << endl;
	  }
};

void imprimirTablaHash(HashTabla<Persona> *ht, string*keys) {
	cout << "SIZE TABLA: " << ht->size()<<endl;
	cout << "SIZE ACTUAL: " << ht->sizeactual() << endl;
	for (int i = 0; i < ht->size(); ++i) {
		cout << ht->buscar(Persona(keys[i],"",0)).getTodo()<< " : " <<keys[i] << endl;
	}
}

int main() {
  auto f_hash=[](Persona p)->int{
    int xd=0;
    for(int i =0; i < p.getNombre().size(); i++){
      int a = p.getNombre()[i];
      xd+=a*27^i;
    }
    return xd;
  };
  auto f_igual=[](Persona p1, Persona p2)->bool{
    return (p1.getNombre().compare(p2.getNombre())==0);
  };
	HashTabla<Persona> *ht = new HashTabla<Persona>(9, f_hash, f_igual);
	string keys[] = {"Will","a","b","c","digito","e","f","g","j"};
	ht->insertar(Persona("Will","Romero",20));
	ht->insertar(Persona("a","Albion",11));
	ht->insertar(Persona("b","Offline",1));
	ht->insertar(Persona("c","Romero",2));
	ht->insertar(Persona("digito","Romero",3));
	ht->insertar(Persona("e","Romero",194));
	ht->insertar(Persona("f","Romero",5));
	ht->insertar(Persona("g","Romero",6));
	ht->insertar(Persona("j","Romero",7));

  //hash = (base + step) % TABLE_SIZE;

	imprimirTablaHash(ht, keys);
	cin.get();

	return 0;
}
```
